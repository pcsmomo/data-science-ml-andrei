```python
from sklearn.model_selection import train_test_split

# Split into X & y
X = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

# Split into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Build machine learning model
from sklearn.ensemble import RandomForestRegressor  # it can predict number

model = RandomForestRegressor()
# model.fit(X_train, y_train)   # ValueError: could not convert string to float: 'Toyota'
# model.score(X_test, y_test)   
```

```python
X.head()
```

```python
# Turn the categories into nubmers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]  # Make: 4, Colour: 5, Doors: 3
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                  one_hot,
                                  categorical_features)],
                                  remainder="passthrough")
transformed_X = transformer.fit_transform(X)
```

```python
pd.DataFrame(transformed_X)
```

```python
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]])
dummies
```

```python
# Let's refit the model
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
